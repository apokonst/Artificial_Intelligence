## Apply transformations to fit the models requirements and store the data

```python
transforms = transforms.Compose([transforms.Resize((232, 232), antialias=True),
                                 transforms.CenterCrop(224),
                                 transforms.ToTensor(),
                                 transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                      std =[0.229, 0.224, 0.225]),])

dataset = datasets.ImageFolder(root="EuroSAT_RGB/EuroSAT_RGB", transform=transforms)
```

```python
# Total images
print("Total images:", len(dataset))

# Class names
print(f"Classes ({len(dataset.classes)}):", dataset.classes)

# Mapping to integer labels
print("Class→index:", dataset.class_to_idx)
```

## Split into training/test

```python
N = len(dataset)
n_train = int(0.8 * N)
n_test = N - n_train

train_ds, test_ds = random_split(dataset, [n_train, n_test])

model_dm = SimpleDataModule(train_ds, test_ds, validation=0.2, batch_size=128)

for idx, (X_ ,Y_) in enumerate(model_dm.train_dataloader()):
    print('X: ', X_.shape)
    print('Y: ', Y_.shape)
    if idx >= 1: # stop after 2 batches
        break
```

## Show some training images

```python
# parameters for “unnormalizing” back to [0,1]
mean = np.array([0.485, 0.456, 0.406])
std  = np.array([0.229, 0.224, 0.225])

# Grab the label list as a NumPy array (fast)
targets = np.array(dataset.targets)   # length 27000

# For each class, find its indices exactly once
label2idx = {cls: np.where(targets == cls)[0]
             for cls in range(len(dataset.classes))}

# Sample one index per class
rng = np.random.default_rng(0)
selected_indices = [rng.choice(label2idx[cls])
                    for cls in range(len(dataset.classes))]

# 4) plot them in 2×5
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for ax, idx in zip(axes.flatten(), selected_indices):
    img_tensor, label = dataset[idx]
    img = img_tensor.numpy().transpose(1, 2, 0)
    img = (img * std) + mean
    img = np.clip(img, 0, 1)
    ax.imshow(img)
    ax.set_title(dataset.classes[label], fontsize=9)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

# **ResNet**

```python
class ResNetClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        # load ImageNet‐pretrained ResNet50
        self.backbone = resnet50(weights=ResNet50_Weights.DEFAULT)

        # Extract number of input features to the final fc layer of the ResNet50 model
        in_feats = self.backbone.fc.in_features

        # Replace the final fc layer with the new one (the desired number of output classes)
        self.backbone.fc = nn.Linear(in_feats, num_classes)

    def forward(self, x):
        return self.backbone(x)
```

```python
resnet_model = ResNetClassifier(num_classes=len(dataset.classes)).to(device)

summary(resnet_model, input_size=(3,224,224))
```

```python
resnet_model.train()
```

## Phase 1: First train only the final layer that we introduced

```python
# freeze everything
for p in resnet_model.parameters():
    p.requires_grad = False

# unfreeze just the new head
for p in resnet_model.backbone.fc.parameters():
    p.requires_grad = True

# Build an optimizer over only the head with a higher LR
head_optimizer_res = RMSprop(resnet_model.backbone.fc.parameters(), lr=1e-2)

# Wrap in SimpleModule and fit
resnet_module = SimpleModule.classification(resnet_model, num_classes=len(dataset.classes),
                                            optimizer=head_optimizer_res)

resnet_trainer1 = Trainer(deterministic=True, max_epochs=5, num_sanity_val_steps = 0,
                         enable_progress_bar=True,
                         callbacks=[ErrorTracker(), progress_bar])

resnet_trainer1.fit(resnet_module, datamodule=model_dm)
```

## Phase 2: Fine-tune the full network

```python
#Un‐freeze everything
for p in resnet_model.parameters():
    p.requires_grad = True

# Re-instantiate optimizer over all parameters at a lower LR
full_optimizer_res = RMSprop(resnet_model.parameters(), lr=1e-4)

# Re-wrap in SimpleModule
resnet_module2 = SimpleModule.classification(resnet_model, num_classes=len(dataset.classes),
                                            optimizer=full_optimizer_res)

# Fit model
resnet_trainer2 = Trainer(deterministic=True, max_epochs=30, num_sanity_val_steps = 0,
                         enable_progress_bar=True,
                         callbacks=[ErrorTracker(), progress_bar])
resnet_trainer2.fit(resnet_module2, datamodule=model_dm)
```

## Test Accuracy (ResNet)

```python
resnet_trainer2.test(resnet_module2, datamodule=model_dm)
```

```python
acc_frac_resnet = resnet_trainer2.logged_metrics["test_accuracy"]
print(f"Test accuracy: {acc_frac_resnet * 100:.1f}%")
```

## Confusion Matrix (ResNet)

```python
# Put your model in eval mode & move to device
resnet_model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
resnet_model.to(device)

# 2) Collect predictions & labels
all_preds, all_labels = [], []
for imgs, labels in model_dm.test_dataloader():
    imgs, labels = imgs.to(device), labels.to(device)
    with torch.no_grad():
        outputs = resnet_model(imgs)
    all_preds.extend(outputs.argmax(dim=1).cpu().numpy())
    all_labels.extend(labels.cpu().numpy())

# 3) Compute normalized confusion matrix
cm = confusion_matrix(all_labels, all_preds, normalize='true')

# 4) Plot
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(cm, cmap='coolwarm')  # default colormap
fig.colorbar(im, ax=ax)
classes = dataset.classes
ax.set_xticks(np.arange(len(classes)))
ax.set_xticklabels(classes, rotation=45, ha='right')
ax.set_yticks(np.arange(len(classes)))
ax.set_yticklabels(classes)
ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
ax.set_title('Normalized Confusion Matrix: ResNet50 on EuroSAT')
for i in range(len(classes)):
    for j in range(len(classes)):
        ax.text(j, i, f"{cm[i,j]:.2f}", ha='center', va='center')
plt.tight_layout()
plt.show()
```

# **AlexNet**

```python
class AlexNetClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        # load the ImageNet‐pretrained AlexNet
        self.backbone = alexnet(weights=AlexNet_Weights.DEFAULT)

        # find how many features its final layer expects:
        in_feats = self.backbone.classifier[6].in_features

        # replace that final Linear so we get `num_classes` outputs
        self.backbone.classifier[6] = nn.Linear(in_feats, num_classes)

    def forward(self, x):
        return self.backbone(x)
```

```python
# instantiate and move to GPU
alex_model = AlexNetClassifier(num_classes=len(dataset.classes)).to(device)
summary(alex_model, input_size=(3,224,224))
```

```python
alex_model.train()
```

## Phase 1: Train only the final layer we introduced

```python
# freeze everything except the new head
for p in alex_model.parameters():
    p.requires_grad = False
for p in alex_model.backbone.classifier[6].parameters():
    p.requires_grad = True

# Build an optimizer over only the head with a higher LR
head_opt_alex = RMSprop(alex_model.backbone.classifier[6].parameters(), lr=1e-2)

# Wrap in SimpleModule and fit
alex_module = SimpleModule.classification(alex_model,
                                          num_classes=len(dataset.classes),
                                          optimizer=head_opt_alex)

alex_trainer1 = Trainer(deterministic=True, max_epochs=5, num_sanity_val_steps=0,
                        enable_progress_bar=True, callbacks=[ErrorTracker(),
                                                             progress_bar])
alex_trainer1.fit(alex_module, datamodule=model_dm)
```

## Phase 2: Fine-tune the whole network

```python
for p in alex_model.parameters():
    p.requires_grad = True

# Re-instantiate optimizer over all parameters at a lower LR
full_opt_alex = RMSprop(alex_model.parameters(), lr=1e-4)

# Re-wrap in SimpleModule
alex_module2 = SimpleModule.classification(alex_model,
                                           num_classes=len(dataset.classes),
                                           optimizer=full_opt_alex)

# Fit the model (deterministic=true does not work with this model)
alex_trainer2 = Trainer(deterministic=False, max_epochs=30, num_sanity_val_steps=0,
                        enable_progress_bar=True,
                        callbacks=[ErrorTracker(), progress_bar])
alex_trainer2.fit(alex_module2, datamodule=model_dm)
```

## Test Accuracy (AlexNet)

```python
alex_trainer2.test(alex_module2, datamodule=model_dm)
```

```python
acc_frac_alex = alex_trainer2.logged_metrics["test_accuracy"]
print(f"Test accuracy: {acc_frac_alex * 100:.1f}%")
```

## Confusion Matrix (AlexNet)

```python
# Put model in eval mode & move to device
alex_model.eval()
alex_model.to(device)

# Collect predictions & labels
all_preds_alex, all_labels_alex = [], []
for imgs_alex, labels_alex in model_dm.test_dataloader():
    imgs_alex, labels_alex = imgs_alex.to(device), labels_alex.to(device)
    with torch.no_grad():
        outputs_alex = alex_model(imgs_alex)
    all_preds_alex.extend(outputs_alex.argmax(dim=1).cpu().numpy())
    all_labels_alex.extend(labels_alex.cpu().numpy())

# 3) Compute normalized confusion matrix
cm_alex = confusion_matrix(all_labels_alex, all_preds_alex, normalize='true')

# 4) Plot
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(cm_alex, cmap='coolwarm')  # default colormap
fig.colorbar(im, ax=ax)
classes = dataset.classes
ax.set_xticks(np.arange(len(classes)))
ax.set_xticklabels(classes, rotation=45, ha='right')
ax.set_yticks(np.arange(len(classes)))
ax.set_yticklabels(classes)
ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
ax.set_title('Normalized Confusion Matrix: AlexNet on EuroSAT')
for i in range(len(classes)):
    for j in range(len(classes)):
        ax.text(j, i, f"{cm_alex[i,j]:.2f}", ha='center', va='center')
plt.tight_layout()
plt.show()
```
